```python
q1, q3 = np.percentile(df['Item_Visibility'], [25, 75])
iqr = q3 -q1

lower_bound, upper_bound = q1 -2.0 * iqr, q3 + 2.0 * iqr

df['Item_Visibility'] = np.where((df['Item_Visibility'] < lower_bound) | (df['Item_Visibility'] > upper_bound), 
                                 np.median(df['Item_Visibility']), df['Item_Visibility'])
```

### Separating independent variable and dependent variable

```python
x2 = quant_df.drop(columns=['Item_Outlet_Sales'])
y2 = quant_df['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split

x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, test_size=0.2, random_state=42)
```

### Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x2_train_scaled = scaler.fit_transform(x2_train)
x2_test_scaled = scaler.transform(x2_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
```

```python
# As features does not follow much linear realtionship between independent 
# variable and target variable we need to add polynomial features.
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(x2_train_scaled)
X_test_poly = poly.transform(x2_test_scaled)
```

```python
lr = LinearRegression()
lr.fit(X_train_poly, y2_train)
```

```python
_pred = lr.predict(X_test_poly)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
R2 = r2_score(y2_test, _pred)
MAE = mean_absolute_error(y2_test, _pred)
MSE = mean_squared_error(y2_test, _pred)
RMSE = np.sqrt(mean_squared_error(y2_test, _pred))
```

```python
print('Scores For Linear regression Model')
print('R2 value :', R2*100)
print('MAE value :', MAE*100)
print('MSE value :', MSE*100)
print('RMSE value :', RMSE*100)
```
